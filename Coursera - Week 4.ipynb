```python
#loading necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook
```

```python
#reading the csv file

url = "https://raw.githubusercontent.com/m23chaffee/DS100-Repository/main/psa%20data%201_sorted.csv"
df = pd.read_csv ('https://raw.githubusercontent.com/m23chaffee/DS100-Repository/main/psa%20data%201_sorted.csv',parse_dates=[2])
df
```

```python
df.columns = df.columns.str.strip()
df.dtypes
```

```python
und = df[['underemployed']]
und
```

```python
unm = df[['unemployed']]
unm
```

```python
df['underemployed']=df['underemployed'].astype(float)
df['unemployed']=df['unemployed'].astype(float)
```

```python
df.info()
```

```python
df.plot()
```

```python
df_new = df.transpose()
```

```python
year_data = df.query("Month == 'Month'")
dfm = df.melt('Month', 
               var_name='Employment Status', 
               value_name = 'Population, per thousand'
              )
dfm
```

```python
#using sns

fig = plt.figure()

sns.lineplot(
    data = dfm,
    x = "Month",
    y = "Population, per thousand",
    hue = 'Employment Status',
    palette = ['#4F1271', '#C8C6D7'],
    markers = True,
    linewidth = 2

    
)

sns.despine(left = True)
plt.xlabel('Month')
plt.ylabel('Population per thousand')
plt.title('2021 Philippine Underemployed and Unemployed Workforce')

plt.gca().spines['bottom'].set_color('#20063B')
plt.gca().spines['bottom'].set_linestyle('-')
plt.gca().yaxis.grid(which = 'major', color = '#000000', linestyle = '-', linewidth = 0.5, alpha = 0.2)
plt.legend(title = 'Employment Status')
```
